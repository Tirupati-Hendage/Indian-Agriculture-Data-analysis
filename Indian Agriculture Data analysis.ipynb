```python
k = crop_production[['crop', 'Production_2006_07', 'Production_2007_08',
       'Production_2008_09', 'Production_2009_10', 'Production_2010_11']].groupby('crop')
index = list(k.indices.keys())
index[-8:-2]
```

```python
k.sum()[-9:-2].plot(figsize=(20,8), kind='bar')
plt.title('Year wise production different agricultural products')
plt.ylabel('Production in Quintal')
```

```python
crop_production.rename(columns = {'Area 2006-07':'Area_2006_07'}, inplace=True)
crop_production.rename(columns = {'Area 2007-08':'Area_2007_08'}, inplace=True)
crop_production.rename(columns = {'Area 2008-09':'Area_2008_09'}, inplace=True)
crop_production.rename(columns = {'Area 2009-10':'Area_2009_10'}, inplace=True)
crop_production.rename(columns = {'Area 2010-11':'Area_2010_11'}, inplace=True)
crop_production.rename(columns = {'Yield_010_1':'Yield_2010_11'}, inplace=True)
crop_production.rename(columns = {'Yield 2006-07':'Yield_2006_07'},inplace=True)
crop_production.rename(columns = {'Yield 2007-08':'Yield_2007_08'},inplace=True)
crop_production.rename(columns = {'Yield_2008-09':'Yield_2008_09'},inplace=True)
crop_production.rename(columns = {'Yield 2009-10':'Yield_2009_10'},inplace=True)
```

```python
total_area = ['Area_2006_07',
              'Area_2007_08',
              'Area_2008_09',
              'Area_2009_10',
              'Area_2010_11',]
```

```python
t1 = crop_production[total_area].sum()

# here we have plotted the graph
plt.title('Total Area Increased')
plt.xticks(rotation=75)
plt.xlabel('Year')
plt.ylabel('Area in Hectare')
plt.plot(t1);
```
